# Warshall algorithm to find reachability between all pairs of vertices in a graph.
Problem Description:
Understand transitive closure of a graph. Use Python to compute reachability using matrix method.

#Theory
Warshall algorithm finds whether a path exists between every pair of vertices. It updates the adjacency matrix step by step. If there is a path from i to j through k, then the entry becomes 1.

## Algorithm
1. Represent graph using adjacency matrix.
2. For each intermediate vertex k
3. Check all pairs i and j
4. Update value if path exists through k
4. Repeat for all vertices


```python
def warshall(graph):
  n=len(graph)
  for k in range(n):
    for i in range(n):
      for j in range(n):
          graph[i][j]=graph[i][j] or (graph[i][k] and graph[k][j])
  return graph
```

```python
graph = [
[0, 1, 1, 0],
[0, 0, 1, 0],
[1, 0, 0, 1],
[0, 0, 0, 0]
]

result = warshall(graph)

print("Transitive Closure Matrix:")

for row in result:
    print(row)
```

### 3. Compare Warshall with Floyd Algorithm

**Warshall's Algorithm (Transitive Closure):**

*   **Purpose:** Determines the reachability between all pairs of vertices in a directed graph. It finds if a path *exists*, not necessarily the shortest one or its cost.
*   **Input:** A boolean adjacency matrix (0s and 1s) representing direct connections.
*   **Output:** A boolean matrix where `result[i][j] = 1` if there is a path from `i` to `j`, and `0` otherwise.
*   **Applicability:** Unweighted graphs, or when only path existence is required.

**Floyd-Warshall Algorithm (All-Pairs Shortest Path):**

*   **Purpose:** Finds the shortest paths (minimum cost/distance) between all pairs of vertices in a weighted graph.
*   **Input:** A weighted adjacency matrix, where `graph[i][j]` is the weight of the edge from `i` to `j`, or `INF` if no direct edge exists.
*   **Output:** A distance matrix where `dist[i][j]` is the length of the shortest path from `i` to `j`.
*   **Applicability:** Weighted graphs (can handle negative weights but not negative cycles), when actual shortest distances are needed.

**Key Differences:**

1.  **Output Type:** Warshall produces a boolean matrix (reachability), while Floyd-Warshall produces a distance matrix (shortest path lengths).
2.  **Input Type:** Warshall typically uses 0/1 for adjacency, while Floyd-Warshall uses edge weights (which can be 0, positive, or negative).
3.  **Information Provided:** Warshall tells you *if* you can get there. Floyd-Warshall tells you *how cheaply/quickly* you can get there.
4.  **Relationship:** Warshall's algorithm can be seen as a special case of Floyd-Warshall where edge weights are either 0 (for direct path existence) or 1 (for an existing path) and the operation `min(A, B)` is replaced by `A or B`, and `A + B` is replaced by `A and B`.

#Understand all pairs shortest path problem. Implement dynamic programming approach in Python.

* Given a weighted graph with positive or negative edge weights, find
shortest distances between every pair of vertices.

# Theory
Floyd Warshall algorithm computes shortest paths between all pairs of nodes. It updates the distance matrix by checking if a shorter path exists through an intermediate vertex. Works with negative weights but fails with negative cycles.

Algorithm
1. Represent graph using distance matrix.
2. Initialize matrix with given weights.
3. Set distance of node to itself as 0.
4. For each intermediate vertex k
5. Update distance[i][j] = min(distance[i][j], distance[i][k] + distance[k][j])
6. Repeat for all vertices

```python
def floyd_warshall(graph):

    n = len(graph)

    dist = []
    for i in range(n):
        dist.append(graph[i][:])

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]

    return dist

```

```python
INF = float('inf')

graph = [
[0, 3, INF, 7],
[8, 0, 2, INF],
[5, INF, 0, 1],
[2, INF, INF, 0]
]

result = floyd_warshall(graph)

print("Shortest distance matrix:")

for row in result:
    print(row)
```

### 2. Detect Negative Cycle in Graph

```python
def floyd_warshall_with_cycle_detection(graph):
    n = len(graph)
    dist = []
    for i in range(n):
        dist.append(graph[i][:])

    # Initialize distances for self-loops to 0
    for i in range(n):
        dist[i][i] = 0

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] != INF and dist[k][j] != INF:
                    dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])

    # Detect negative cycles
    has_negative_cycle = False
    for k in range(n):
        for i in range(n):
            for j in range(n):

                if dist[i][k] != INF and dist[k][j] != INF and dist[i][k] + dist[k][j] < dist[i][j]:

                    has_negative_cycle = True

    for i in range(n):
        if dist[i][i] < 0:
            has_negative_cycle = True
            break # No need to check further if one is found

    return dist, has_negative_cycle

shortest_distances, has_neg_cycle = floyd_warshall_with_cycle_detection(user_graph)

print("\nShortest distance matrix (with user input):")
for row in shortest_distances:
    print([val if val != INF else 'INF' for val in row])

if has_neg_cycle:
    print("\nGraph contains a negative cycle.")
else:
    print("\nGraph does NOT contain a negative cycle.")
```

### 3. Print Shortest Path Between Two Given Vertices

```python
def floyd_warshall_with_paths(graph):
    n = len(graph)
    dist = []
    path = [] # Predecessor matrix

    for i in range(n):
        dist.append(graph[i][:])
        path.append([None] * n)
        for j in range(n):
            if i == j:
                path[i][j] = i # Or None, depends on convention. Here, self-loop has no intermediate.
            elif graph[i][j] != INF:
                path[i][j] = i # Direct edge, predecessor is i

    # Initialize distances for self-loops to 0
    for i in range(n):
        dist[i][i] = 0

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] != INF and dist[k][j] != INF:
                    if dist[i][k] + dist[k][j] < dist[i][j]:
                        dist[i][j] = dist[i][k] + dist[k][j]
                        path[i][j] = path[k][j] # Update predecessor through k

    # Detect negative cycles
    has_negative_cycle = False
    for i in range(n):
        if dist[i][i] < 0:
            has_negative_cycle = True
            break

    return dist, path, has_negative_cycle

def reconstruct_path(path_matrix, start_node, end_node):
    if path_matrix[start_node][end_node] is None:
        return [] # No path

    current = end_node
    path = [current]
    while current != start_node:
        if path_matrix[start_node][current] is None: # Should not happen if path exists, but as a safeguard
             return [] # Path not reconstructable
        current = path_matrix[start_node][current]
        path.append(current)
    return path[::-1] # Reverse to get path from start to end

shortest_distances, path_matrix, has_neg_cycle = floyd_warshall_with_paths(user_graph)

print("\nShortest distance matrix (with user input) after path calculation:")
for row in shortest_distances:
    print([val if val != INF else 'INF' for val in row])

if has_neg_cycle:
    print("\nGraph contains a negative cycle. Paths might not be meaningful.")
else:
    print("\nGraph does NOT contain a negative cycle.")

# Get user input for start and end vertices
while True:
    try:
        start_vertex = int(input(f"Enter the start vertex (0 to {len(user_graph)-1}): "))
        end_vertex = int(input(f"Enter the end vertex (0 to {len(user_graph)-1}): "))
        if 0 <= start_vertex < len(user_graph) and 0 <= end_vertex < len(user_graph):
            break
        else:
            print("Invalid vertex number. Please try again.")
    except ValueError:
        print("Invalid input. Please enter an integer.")

if shortest_distances[start_vertex][end_vertex] == INF:
    print(f"\nNo path exists from {start_vertex} to {end_vertex}.")
else:
    print(f"\nShortest distance from {start_vertex} to {end_vertex}: {shortest_distances[start_vertex][end_vertex]}")
    path = reconstruct_path(path_matrix, start_vertex, end_vertex)
    if path:
        print(f"Shortest path: {' -> '.join(map(str, path))}")
    else:
        print("Could not reconstruct path.")
```

### 4. Compare with Dijkstra Algorithm

```python
import heapq

def dijkstra(graph, start_node):
    n = len(graph)
    distances = [INF] * n
    distances[start_node] = 0
    priority_queue = [(0, start_node)] # (distance, vertex)

    while priority_queue:
        dist, u = heapq.heappop(priority_queue)

        if dist > distances[u]:
            continue

        for v in range(n):
            if graph[u][v] != INF and distances[u] + graph[u][v] < distances[v]:
                distances[v] = distances[u] + graph[u][v]
                heapq.heappush(priority_queue, (distances[v], v))

    return distances

print("\nDijkstra's Algorithm Results (for each starting node):")
for start_node_dijkstra in range(len(user_graph)):
    dijkstra_distances = dijkstra(user_graph, start_node_dijkstra)
    print(f"From node {start_node_dijkstra}: {[val if val != INF else 'INF' for val in dijkstra_distances]}")
```

## Comparison: Floyd-Warshall vs. Dijkstra

**Floyd-Warshall Algorithm:**
*   **Type:** All-Pairs Shortest Path (finds shortest paths between all pairs of vertices).
*   **Approach:** Dynamic programming. Iteratively updates the shortest path between `i` and `j` considering `k` as an intermediate vertex.
*   **Time Complexity:** O(V^3), where V is the number of vertices. This makes it suitable for dense graphs or when all-pairs shortest paths are needed.
*   **Negative Weights:** Handles negative edge weights correctly.
*   **Negative Cycles:** Can detect negative cycles, but if one exists, the shortest paths are undefined for affected pairs.
*   **Implementation:** Simpler to implement due to its iterative structure.

**Dijkstra's Algorithm:**
*   **Type:** Single-Source Shortest Path (finds shortest paths from a single source vertex to all other vertices).
*   **Approach:** Greedy algorithm. Builds the shortest path tree from the source by iteratively selecting the unvisited vertex with the smallest distance.
*   **Time Complexity:** O(E log V) or O(E + V log V) with a Fibonacci heap, or O(V^2) with an adjacency matrix and simple array for minimum extraction. E is the number of edges. More efficient for sparse graphs or when only single-source shortest paths are needed.
*   **Negative Weights:** Does NOT handle negative edge weights correctly (will produce incorrect results if negative weights are present).
*   **Negative Cycles:** Not designed to detect negative cycles and will likely loop indefinitely or produce incorrect results if one is present.
*   **Implementation:** Requires a priority queue, making implementation slightly more complex.

**When to use which:**
*   **Floyd-Warshall:** Use when you need the shortest path between *all* pairs of vertices, or when the graph has negative edge weights (but no negative cycles), or if the graph is dense (E approaches V^2).
*   **Dijkstra:** Use when you need the shortest paths from a *single* source vertex to all other vertices, and the graph has *non-negative* edge weights. It is generally more efficient for sparse graphs in this scenario.
